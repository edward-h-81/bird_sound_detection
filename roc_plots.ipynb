```python
import csv
import matplotlib.pyplot as plt

# data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_ff_test.csv'
# data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_warblr_test.csv'
data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_birdvox_test.csv'

baseline_fpr = []
baseline_tpr = []
bg_fpr = []
bg_tpr = []
cs_fpr = []
cs_tpr = []
pink_fpr = []
pink_tpr = []
tf_fpr = []
tf_tpr = []
ensemble_fpr = []
ensemble_tpr = []

with open(data, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr.append(float(line[0]))
      baseline_tpr.append(float(line[1]))
    if line[2] != '':
      bg_fpr.append(float(line[2]))
      bg_tpr.append(float(line[3]))
    if line[4] != '':
      cs_fpr.append(float(line[4]))
      cs_tpr.append(float(line[5]))
    if line[6] != '':
      pink_fpr.append(float(line[6]))
      pink_tpr.append(float(line[7]))
    if line[8] != '':
      tf_fpr.append(float(line[8]))
      tf_tpr.append(float(line[9]))
    if line[10] != '':
      ensemble_fpr.append(float(line[10]))
      ensemble_tpr.append(float(line[11]))
    

plt.figure(1)
plt.figure(figsize=(9.7,9.7))
plt.plot([0, 1], [0, 1], 'y--')
plt.plot(baseline_fpr, baseline_tpr,   linewidth=0.8)
plt.plot(bg_fpr, bg_tpr,  linewidth=0.5, linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round',)
plt.plot(cs_fpr, cs_tpr, linewidth=0.5)
plt.plot(pink_fpr, pink_tpr,  linewidth=0.5, linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round',)
plt.plot(tf_fpr, tf_tpr,  linewidth=0.5)
plt.plot(ensemble_fpr, ensemble_tpr,   linewidth=0.8)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve: architecture trained on FreeField and Warblr; tested on BirdVox')
plt.legend(["Random", "Baseline", "Background Noise", "Cyclic Time Shift", "Pink Noise", "Time-Freq Mask", "Ensemble"], loc=4)
plt.show
```

Baseline and ensemble models plotted

```python
import csv
import matplotlib.pyplot as plt

# data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_ff_test.csv'
# data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_warblr_test.csv'
data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_birdvox_test.csv'

baseline_fpr = []
baseline_tpr = []

ensemble_fpr = []
ensemble_tpr = []

with open(data, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr.append(float(line[0]))
      baseline_tpr.append(float(line[1]))

    if line[10] != '':
      ensemble_fpr.append(float(line[10]))
      ensemble_tpr.append(float(line[11]))
    

plt.figure(1)
plt.figure(figsize=(9.7,9.7))
plt.plot([0, 1], [0, 1], 'y--')
plt.plot(baseline_fpr, baseline_tpr,   linewidth=1.8, linestyle=(0, (5, 2, 1, 2)))
plt.plot(ensemble_fpr, ensemble_tpr,   linewidth=1.8, linestyle=(2, (5, 2, 1, 2)))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve: architecture trained on Freefield and Warblr; tested on BirdVox')
plt.legend(["Random", "Baseline", "Ensemble"], loc=4)
plt.show
```

Models trained and tested on different data

```python
import csv
import matplotlib.pyplot as plt

data_ff = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_ff_test.csv'
data_data_warblr = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_warblr_test.csv'
data_bv = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_birdvox_test.csv'

baseline_fpr_warblr = []
baseline_tpr_warblr = []
ensemble_fpr_warblr = []
ensemble_tpr_warblr = []

baseline_fpr_ff = []
baseline_tpr_ff = []
ensemble_fpr_ff = []
ensemble_tpr_ff = []

baseline_fpr_bv = []
baseline_tpr_bv = []
ensemble_fpr_bv = []
ensemble_tpr_bv = []

with open(data_data_warblr, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr_warblr.append(float(line[0]))
      baseline_tpr_warblr.append(float(line[1]))

    if line[10] != '':
      ensemble_fpr_warblr.append(float(line[10]))
      ensemble_tpr_warblr.append(float(line[11]))

with open(data_ff, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr_ff.append(float(line[0]))
      baseline_tpr_ff.append(float(line[1]))

    if line[10] != '':
      ensemble_fpr_ff.append(float(line[10]))
      ensemble_tpr_ff.append(float(line[11]))

with open(data_bv, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr_bv.append(float(line[0]))
      baseline_tpr_bv.append(float(line[1]))

    if line[10] != '':
      ensemble_fpr_bv.append(float(line[10]))
      ensemble_tpr_bv.append(float(line[11]))
    

plt.figure(1)
plt.figure(figsize=(9.7,9.7))
plt.plot([0, 1], [0, 1], 'y--')
plt.plot(baseline_fpr_warblr, baseline_tpr_warblr,   linewidth=1.0)
plt.plot(ensemble_fpr_warblr, ensemble_tpr_warblr,   linewidth=1.4, linestyle=(2, (5, 2, 1, 2)))
plt.plot(baseline_fpr_ff, baseline_tpr_ff,   linewidth=1.0)
plt.plot(ensemble_fpr_ff, ensemble_tpr_ff,   linewidth=1.4, linestyle=(2, (5, 2, 1, 2)))
plt.plot(baseline_fpr_bv, baseline_tpr_bv,   linewidth=1.0)
plt.plot(ensemble_fpr_bv, ensemble_tpr_bv,   linewidth=1.4, linestyle=(1, (5, 2, 1, 2)), color="#F50B94")
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve: models trained and tested on different folds')
plt.legend(["Random", "Baseline BirdVox+Freefield", "Ensemble BirdVox+Freefield", 
            "Basline BirdVox+Warblr", "Ensemble BirdVox+Warblr",
            "Baseline FreeField+Warblr", "Ensemble FreeField+Warblr"], loc=4)
plt.show

# plt.savefig('/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_curve.png')
```
