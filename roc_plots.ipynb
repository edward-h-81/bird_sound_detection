```python
import csv
import matplotlib.pyplot as plt

data = '/content/drive/My Drive/MSc_Project_Colab/BAD_PyTorch/roc_data_ff_test.csv'

baseline_fpr = []
baseline_tpr = []
bg_fpr = []
bg_tpr = []
cs_fpr = []
cs_tpr = []
pink_fpr = []
pink_tpr = []
tf_fpr = []
tf_tpr = []

with open(data, 'r') as csv_file:
  csv_reader = csv.reader(csv_file)

  next(csv_reader)

  for line in csv_reader:
    if line[0] != '':
      baseline_fpr.append(float(line[0]))
      baseline_tpr.append(float(line[1]))
    if line[2] != '':
      bg_fpr.append(float(line[2]))
      bg_tpr.append(float(line[3]))
    if line[4] != '':
      cs_fpr.append(float(line[4]))
      cs_tpr.append(float(line[5]))
    if line[6] != '':
      pink_fpr.append(float(line[6]))
      pink_tpr.append(float(line[7]))
    if line[8] != '':
      tf_fpr.append(float(line[8]))
      tf_tpr.append(float(line[9]))
    

plt.figure(1)
plt.figure(figsize=(9.7,9.7))
plt.plot([0, 1], [0, 1], 'y--')
plt.plot(baseline_fpr, baseline_tpr,  linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.plot(bg_fpr, bg_tpr, linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.plot(cs_fpr, cs_tpr,linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.plot(pink_fpr, pink_tpr, linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.plot(tf_fpr, tf_tpr, linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(["Random", "Baseline", "Background Noise", "Cyclic Time Shift", "Pink Noise", "Time-Freq Mask"], loc=4)
plt.show
```
